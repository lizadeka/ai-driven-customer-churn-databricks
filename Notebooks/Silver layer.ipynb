```python
%sql
DESCRIBE bronze_telco_churn;
```

```python
from pyspark.sql.functions import col, trim, when

silver_df = (
    spark.table("bronze_telco_churn")
    .withColumn(
        "TotalCharges",
        when(trim(col("TotalCharges")) == "", None)
        .otherwise(col("TotalCharges"))
        .cast("double")
    )
    .withColumn(
        "SeniorCitizen",
        when(col("SeniorCitizen") == 1, "Yes")
        .otherwise("No")
    )
)
```

```python
silver_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("silver_telco_churn")
```

```python
%sql
SELECT
    COUNT(*) AS null_total_charges
FROM silver_telco_churn
WHERE TotalCharges IS NULL;
```
